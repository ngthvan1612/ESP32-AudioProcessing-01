```python
import numpy as np
import scipy
import os
import IPython
```

```python
INPUT_FILE_NAME = './data/trai-tien-lui-phai-01.wav'
```

```python
sample_rate, audio_frame = scipy.io.wavfile.read(INPUT_FILE_NAME)
```

```python
audio_frame = np.array(audio_frame)
```

```python
print("Audio length (in seconds) = ", audio_frame.shape[0] / sample_rate)
```

```python
def show_video_frame(source, start, end):
  start = int(start)
  end = int(end)
  scipy.io.wavfile.write("./temp/test.wav", sample_rate, source[start:end])
  IPython.display.display(
    IPython.display.Audio('./temp/test.wav')
  )
  return source[start:end]
```

```python
trai_01 = show_video_frame(audio_frame, sample_rate * 3.4, sample_rate * 4.0)
```

```python
trai_02 = show_video_frame(audio_frame, sample_rate * 4.4, sample_rate * 5.0)
```

```python
tien_01 = show_video_frame(audio_frame, sample_rate * 7.5, sample_rate * 8.1)
```

```python
tien_02 = show_video_frame(audio_frame, sample_rate * 9.2, sample_rate * 9.8)
```

```python
lui_01 = show_video_frame(audio_frame, sample_rate * 10.4, sample_rate * 11.0)
```

```python
lui_02 = show_video_frame(audio_frame, sample_rate * 12.2, sample_rate * 12.8)
```

```python
phai_01 = show_video_frame(audio_frame, sample_rate * 13.6, sample_rate * 14.2)
```

```python
phai_02 = show_video_frame(audio_frame, sample_rate * 15.0, sample_rate * 15.6)
```

```python
raw_data = [
  trai_01,
  trai_02,
  tien_01,
  tien_02,
  lui_01,
  lui_02,
  phai_01,
  phai_02
]

raw_label = [
  0,
  0,
  1,
  1,
  2,
  2,
  3,
  3
]
```
